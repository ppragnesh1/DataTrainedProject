### Null values in each columns by Index number and Mode value of that column

```python
for i in df.columns[1:-1]:
    if df[i].isnull().sum()>0:
        print("*"*100)
        print(i, "list of index number with Null value", df[df[i].isnull()][i].index.to_list())
        print(i," Mode Value in this column ",df[i].mode()[0])
        print("\n")
    else:
        print(i, "has Zero null values\n")
```

### Filling the Null values :

```python
# finding the total count of null values in each columns
df.isnull().sum().to_frame()
```

```python
dfmode = df.copy(deep=True)
```

```python
for i in dfmode.columns[1:-1]:
    if df[i].isnull().sum()>0:
        print("*"*100)
        clmmode = dfmode[i].mode()[0]
        print(i," = Mode is =",clmmode)
        print(dfmode[i].isnull().sum())
        dfmode[i].fillna(clmmode,inplace=True)
        print("After fill a null value : ",dfmode[i].isnull().sum(),"\n")
        print("*"*100)
    else:
        print(f"Non Null value in :  {i}\n ")
```

### Scaling the data using Min-Max Scaler:

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
lr = LinearRegression()
```

```python
max_train=0
max_test=0
R_state = 0
for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    trainscore = r2_score(y_train,pred_train)
    testscore = r2_score(y_test,pred_test)
    if testscore>max_test:
        max_test=testscore
        max_train=trainscore
        R_state=i
        print("*"*50)
        print("At Random State : ",R_state)
        print("Train Score :",max_train)
        print("Test score : ",max_test)
        print("Differnce :",max_train-max_test)
        
    
```

At Random State "0" we are getting Test score : 0.95 with lowest difference between the train and test.

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

# Train the Model

```python
lr.fit(x_train,y_train)
```

```python
pred_test = lr.predict(x_test)
pred_train = lr.predict(x_train)
print(r2_score(y_test,pred_test)*100)
```

### Cross Validation of the model

```python
Train_accuracy = r2_score(y_train,pred_train)
Test_accuracy = r2_score(y_test,pred_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for j in range(2,20):
    cv_score = cross_val_score(lr,x,y,cv=j)
    cv_mean = cv_score.mean()
    print(f"At Cross fold {j} the cv score is {cv_mean}")
    print("Accuracy score for Training is",Train_accuracy)
    print("Accuracy score for Testing is",Test_accuracy)
    print("Differnce is :" ,Test_accuracy-cv_mean)
    print("\n","*"*50)
```

Cross fold 19 giving a good cross val score of 0.9120 with very less difference of 0.0390

```python
# Cross val score for linear regression

cv_score = cross_val_score(lr,x,y,cv=19)
cv_mean_lr = cv_score.mean()
print("LR CV score is :",cv_mean_lr)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
from sklearn.model_selection import KFold
from sklearn.linear_model import ElasticNet
```

```python
parameters = {"alpha":[.0001,0.001,0.01,.1,1,10],
             "random_state":list(range(0,10))}
folds = KFold(n_splits=5,shuffle = True,random_state=0)

ls = Lasso()
clf=GridSearchCV(ls,parameters,scoring="r2",cv=folds,return_train_score=True,verbose=1)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
ls = Lasso(alpha=0.01,random_state=0)
ls.fit(x_train,y_train)
ls_score_train = ls.score(x_train,y_train)
pred_ls = ls.predict(x_test)
print("Lasso :",ls_score_train*100)
```

```python
# Cross val score 
cv_score = cross_val_score(ls,x,y,cv=19)
cv_mean = cv_score.mean()
print("Lass Cross Val score : ",cv_mean)
```

```python
en =  ElasticNet()

model_cv_en = GridSearchCV(estimator=en,param_grid=parameters,scoring="r2",cv=folds,return_train_score=True,verbose=1)
model_cv_en.fit(x_train,y_train)
print(model_cv_en.best_params_)
```

```python
en = ElasticNet(alpha=0.01,random_state=0)
en.fit(x_train,y_train)

#predit
pred_train= en.predict(x_train)
print("EN Train Score",r2_score(y_train,pred_train))

pred_test = en.predict(x_test)
print("EN Test score",r2_score(y_test,pred_test))
```

```python
# Cross val score 
cv_score = cross_val_score(en,x,y,cv=19)
cv_mean = cv_score.mean()
print("EN Cross Val score : ",cv_mean)
```

### Ensemble Technique

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters = {"criterion": ["mse","mae"],
             "max_features":["auto","sqrt","log2"]}

rf = RandomForestRegressor()
clf_rf = GridSearchCV(rf,parameters)
clf_rf.fit(x_train,y_train)
print(clf_rf.best_params_)
```

```python
rf = RandomForestRegressor(criterion="mse",max_features="log2")
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_rf=rf.predict(x_test)

print("RF Testing Score",r2_score(y_test,pred_rf))
```

```python
cv_score = cross_val_score(rf,x,y,cv=19)
cv_mean_rf = cv_score.mean()
print("RF cross val score :",cv_mean_rf)
```

### Lasso model has lowest difference between Test accuracy score and cross val score with hyper parameter tuning. so we will select the final model as lasso
