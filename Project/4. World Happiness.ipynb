## Separating the Columns into Features and Target

```python
x = dfnew.iloc[:,1:]
y = dfnew.iloc[:,:1]
```

```python
x.shape
```

```python
y.shape
```

## Scaling the data unsing Min-Max Scaler

```python
# from sklearn.preprocessing import MinMaxScaler
# mms = MinMaxScaler()
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_sc = sc.fit_transform(x)
```

```python
x_sc
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x_sc,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    print("At Random State :", i)
    print("Training Accurcy Score :",r2_score(y_train,pred_train)*100)
    print("Testing Accuracy Score :",r2_score(y_test,pred_test)*100)
    print("Difference :", (r2_score(y_train,pred_train)*100)-(r2_score(y_test,pred_test)*100))
    print("\n")
    
```

At Random State  0 with diff of Accuracy train and test : 0.64
At Random State 52 with diff of Accuracy Train and Test : 0.35
At Random State 24 with diff of Accuracy Train and TEst :-0.43    

so we will select Random State : 24

# Linear Regression Model

```python
x_train,x_test,y_train,y_test=train_test_split(x_sc,y,test_size=0.2,random_state=52)
lr.fit(x_train,y_train)
pred_train=lr.predict(x_train)
pred_test=lr.predict(x_test)
print("LR - Accuracy Score ",r2_score(y_test,pred_test)*100)
```

```python
# Checking cross _val _score
train_accuracy = r2_score(y_train,pred_train)
test_accuracy = r2_score(y_test,pred_test)

from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cvscore = cross_val_score(lr,x_sc,y,cv=j).mean()
    print("At Cross fold :", j)
    print("C V Score :",cvscore)
    print("Train Accuracy :",train_accuracy)
    print("Test Accuracy : ",test_accuracy)
    print("\n")
```

Lowest differnce we can see between CV score and Test Accuracy is at CV:6

## Checking more models 

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor
from sklearn.linear_model import SGDRegressor, LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor

from sklearn.model_selection import cross_validate, cross_val_score
from sklearn.metrics import explained_variance_score, max_error, mean_absolute_error, r2_score, explained_variance_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error, mean_squared_error, mean_squared_log_error
from sklearn.metrics import median_absolute_error, mean_poisson_deviance, mean_gamma_deviance
from sklearn.model_selection import KFold
```

```python
models=[("Linear Regression", LinearRegression()),
        ("Ridge Regression", Ridge()),
        ("Lasso Regression", Lasso()),
        ("Stochastic Gradient Descent", SGDRegressor()),
        ("Decision Tree", DecisionTreeRegressor()),
        ("Random Forest", RandomForestRegressor()),
        ("Gradient Boosting", GradientBoostingRegressor()),
        ("Ada Boost", AdaBoostRegressor()),
        ("XGBRegressor", XGBRegressor()),
        ("KNeighbors", KNeighborsRegressor()),
        ("SVM linear", SVR(kernel='linear')),
        ("SVM rbf", SVR(kernel='rbf'))]


for name,model in models:
    model.fit(x_train,y_train)
    modelpredict=model.predict(x_test)
    print("Model :",name)
    print("score: ", model.score(x_test, y_test))
    print("cross_val_score: ", cross_val_score(model,x_train, y_train, cv = 10).mean())
    print("r2_score: ", r2_score(y_test, modelpredict))
    print("")
    print("mean_absolute_error: ", mean_absolute_error(y_test, modelpredict))
    print("mean_squared_error: ", mean_squared_error(y_test, modelpredict))
    print("root_mean_squared_error: ", mean_squared_error(y_test, modelpredict, squared=False))
    print("max_error: ", max_error(y_test, modelpredict))
    print("========================\n")
```

# Hyper parameter testing

```python
from sklearn.model_selection import GridSearchCV

parameters = { 'loss' : ['ls', 'lad', 'huber', 'quantile'],
              'learning_rate' : (0.05,0.25,0.50,1),
              'criterion' : ['friedman_mse', 'mse', 'mae'],
              'max_features' : ['auto', 'sqrt', 'log2']
             }
grid = GridSearchCV(GradientBoostingRegressor(),parameters)
model = grid.fit(x_sc,y)
print(model.best_params_,"\n")
print(model.best_estimator_,"\n")
```

```python
# Checking the GradientBoostingRegressior

FinalModel = GradientBoostingRegressor(learning_rate=0.05, loss='lad', max_features='log2')

FinalModel.fit(x_train,y_train)
modelpredict=FinalModel.predict(x_test)
print("score: ", FinalModel.score(x_test, y_test))
print("cross_val_score: ", cross_val_score(FinalModel,x_train, y_train, cv = 10).mean())
print("r2_score: ", r2_score(y_test, modelpredict))
print("")
print("mean_absolute_error: ", mean_absolute_error(y_test, modelpredict))
print("mean_squared_error: ", mean_squared_error(y_test, modelpredict))
print("root_mean_squared_error: ", mean_squared_error(y_test, modelpredict, squared=False))
print("max_error: ", max_error(y_test, modelpredict))
print("========================\n")
```
